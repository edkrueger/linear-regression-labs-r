## Validation Set Approach (Instructor)

```r
# make a function to split the data into a training set and a validation set
test_train_split = function(data, test_size=.25){
    
    # the the number of observations in the data
    n_obs = dim(mtcars)[1]
    
    # find the number of training observations by rounding up
    n_train_obs = ceiling(n_obs * (1 - test_size))
    
    # get the train indexes
    train_idx = sample(n_obs, n_train_obs)
    
    # split the data
    train = data[train_idx, ]
    test = data[-train_idx, ]

    # return the result
    result = list(train, test)
    return(result)
}
```

```r
# write a function to calculate the MSE
calc_MSE = function (model, data){
    return (
        mean(
            (data$mpg - predict(model, data)) ^ 2
        )
    )
}
```

```r
# split the data
set.seed(1)
result = test_train_split(mtcars)
train_data = data.frame(result[1])
test_data = data.frame(result[2])
train_data
test_data
```

```r
# train some models
lm_1 = lm(mpg ~ hp, data=train_data)
lm_2 = lm(mpg ~ poly(hp, 2), data=train_data)
lm_3 = lm(mpg ~ poly(hp, 3), data=train_data)
```

```r
# calculate the train MSE for the three models
calc_MSE(lm_1, train_data)
calc_MSE(lm_2, train_data)
calc_MSE(lm_3, train_data)
```

```r
# calculate the test MSE for the three models
calc_MSE(lm_1, test_data)
calc_MSE(lm_2, test_data)
calc_MSE(lm_3, test_data)
```

```r
set.seed(2)
result = test_train_split(mtcars)
train_data = data.frame(result[1])
test_data = data.frame(result[2])
train_data
test_data
```

```r
# train the models again on the new train data
lm_1 = lm(mpg ~ hp, data=train_data)
lm_2 = lm(mpg ~ poly(hp, 2), data=train_data)
lm_3 = lm(mpg ~ poly(hp, 3), data=train_data)
```

```r
# calculate the train MSE for the three models
calc_MSE(lm_1, train_data)
calc_MSE(lm_2, train_data)
calc_MSE(lm_3, train_data)
```

```r
# calculate the test MSE for the three models
calc_MSE(lm_1, test_data)
calc_MSE(lm_2, test_data)
calc_MSE(lm_3, test_data)
```

```r
# the the glm function has the same behavior as the lm function by default, but with expanded output
# we'll use the glm because cv.glm, the function in boot, works with glm objects

lm_1 = glm_1 = glm(mpg ~ hp, data=mtcars)
glm_1 = glm(mpg ~ hp, data=mtcars)

summary(lm_1)
summary(glm_1)
```

```r
# cross-validate the linear through 10th degree model of horsepower on mpg

# set a seed
set.seed(17)

# create a vector of 10 0s to store the results
cv_errors = rep(0, 10)

for (i in 1:10) {
    glm_fit = glm(mpg ~ poly(hp, i), data=mtcars)
    cv_errors[i] = cv.glm(mtcars, glm_fit, , K=10)$delta[1]
}

cv_errors
```

## Best Subset Selection (Instructor)

```r
# load ISLR for the auto data set
# since ISLR isn't available in a conda channel, you'll have to build it
# see `install-instructions/README.md`
library(ISLR)
```

```r
# rename the dataset's variable and see some information about the dataset
auto = Auto
names(auto)
dim(auto)
```

```r
# convert origin to a factor
auto$origin = factor(auto$origin)
```

```r
# see the first 10 rows to identify duplicate names
head(auto[order(auto$name),], 10)
```

```r
# create a new column with the name and year concatenated
auto$name_year = paste(auto$name, auto$year)
```

```r
# see all of the duplicated name_years
auto[duplicated(auto$name_year) | duplicated(auto$name_year, fromLast=TRUE),] 
```

```r
# keep only the first duplicate, drop the rest
auto = auto[!duplicated(auto$name_year),]
```

```r
# set name_year as the index
rownames(auto) = auto$name_year
auto$name_year = NULL
auto$name= NULL
head(auto)
```

```r
# plot the pairwise scatter plots
plot(auto)
```

```r
# load the leaps library which is available in conda's r channel
# see the instructions in install-instructions/README.md
library(leaps)
```

```r
# use subset selection to find the best model for mpg for every number of variables

to = proc.time()

reg_fit_full = regsubsets(mpg ~ ., data=auto)
tf = proc.time()
tf - to

reg_summary = summary(reg_fit_full)
reg_summary
```

```r
# see the R-squared for the best model with idx variables
reg_summary$rsq
```

```r
# plot number of variables for the best model against the associated RSS, Adjusted R-Squared, Cp and BIC. 
par(mfrow=c(2, 2))

plot(reg_summary$rss, xlab="Number of Variables", ylab="RSS", type="l")
points(
    which.min(reg_summary$rss),
    min(reg_summary$rss),
    col="red",
    cex=2,
    pch=20
)

plot(reg_summary$adjr2, xlab="Number of Variables", ylab="Adjusted RSq", type="l")
points(
    which.max(reg_summary$adjr2),
    max(reg_summary$adjr2),
    col="red",
    cex=2,
    pch=20
)

plot(reg_summary$cp, xlab="Number of Variables", ylab="Cp", type="l")
points(
    which.min(reg_summary$cp),
    min(reg_summary$cp),
    col="red",
    cex=2,
    pch=20
)

plot(reg_summary$bic, xlab="Number of Variables", ylab="BIC", type="l")
points(
    which.min(reg_summary$bic),
    min(reg_summary$bic),
    col="red",
    cex=2,
    pch=20
)
```

```r
# use the built-in plot function to plot the best models against their associated fit metrics
par(mfrow = c(2,2))
plot(reg_fit_full, scale="r2")
plot(reg_fit_full, scale="adjr2")
plot(reg_fit_full, scale="Cp")
plot(reg_fit_full, scale="bic")
```

```r
# see the coefficients for the best model with 4 variable
coef(reg_fit_full, 4)
```
